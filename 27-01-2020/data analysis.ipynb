```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
sns.set(style="darkgrid")
t1=sns.load_dataset("tips")
t1.tail() #head()
```

```python
sns.relplot(x="total_bill",y="day",data=t1)
```

```python
sns.relplot(x="total_bill",y="day",data=t1,hue="tip")
```

```python
sns.relplot(x="total_bill",y="size",data=t1,hue="tip",kind="line") #line graph
```

```python
sns.set(style="ticks")
t=sns.load_dataset("flights")
t.head()
```

```python
sns.relplot(x="year",y="passengers",hue="month",data=t)
```

```python
sns.relplot(x="year",y="passengers",hue="month",col="month",data=t)
```

```python
sns.set(style="whitegrid")
i=sns.load_dataset("iris")
i.head()
```

```python
i["species"].value_counts()
```

```python
sns.relplot(x="sepal_width",y="petal_width",data=i)
```

```python
sns.relplot(x="sepal_width",y="petal_width",data=i,hue="species",col="species")
```

```python
sns.relplot("total_bill","tip",data=t1,
            style="smoker",size="size",
            col="time")
```

```python
t1.tail()
```

```python
sns.lmplot("total_bill","tip",hue="smoker",data=t1,col="time")
```

```python
#categorial data plot
#catplot
#swarm plot
#scatter plot
#bar plot
```

```python
sns.catplot(x="species",y="petal_length",data=i)
```

```python
sns.catplot(x="species",y="petal_length",
            hue="sepal_width",data=i,kind="swarm")
```

```python
sns.catplot(x="species",y="petal_length",
            hue="sepal_width",data=i,kind="violin")
```

```python
#box plot
f,axs=plt.subplots(1,2)
sns.boxplot(x="species",y="sepal_length",data=i,ax=axs[0])
sns.scatterplot(x="species",y="sepal_length",data=i,ax=axs[1])
plt.show
```

```python
sns.swarmplot(x="total_bill",y="day",data=t1)
```

```python
#visualization of dataset
#joinplot
#kde-kernel density
#hex
#scatter
```

```python
sns.jointplot(x="total_bill",y="tip",data=t1,kind="kde",hue="smoker")
```

```python
sns.jointplot(x="total_bill",y="tip"
              ,data=t1,kind="hex")
```

```python
sns.jointplot(x="total_bill",y="tip"
              ,data=t1,kind="reg",color='g',height=3.5,ratio=5)
```

```python
#pairplot
sns.pairplot(data=i)
```

```python
sns.pairplot(data=i,hue="species",palette="husl")
```

```python
sns.pairplot(data=i,hue="species",markers=["o","s","D"])
```

```python
sns.pairplot(i,vars=["sepal_width","sepal_length"])
,
                     #"petal_width","petal_length"])
```

```python
sns.pairplot(i,diag_kind="kde",markers="+")
```

```python
#distplot
x=np.random.rand(50)
sns.distplot(x)
```

```python
#facetgrid plot
ff=sns.FacetGrid(i,col="species")
ff.map(plt.hist,"sepal_length")
```
