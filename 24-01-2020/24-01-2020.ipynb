```python
# creating subplot using numpy with sine
import numpy as np
t=np.arange(0.0,0.5,0.1)
s=np.sin(2*np.pi*t)
plt.subplot(2,2,3)
plt.plot(t,s,)
plt.xlabel("t")
plt.ylabel("s")
plt.title("graph t vs s")
plt.legend()
plt.show()
```

```python
import numpy as np
t=np.arange(0.0,0.5,0.1)
s=np.exp(-t)*np.sin(2*np.pi*t)
plt.subplot(2,2,3)
plt.plot(t,s,label="1st graph with sine wave")
```

# hist, scatter using subplot

```python
fig=plt.figure()
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(222)
plt.plot(np.random.rand(39))
ax1.hist(np.random.rand(50),bins=20,color='r')
ax2.scatter(np.arange(30),np.arange(30)*8)
```
